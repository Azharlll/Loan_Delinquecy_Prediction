```python
# importing python libraries
import numpy as np
import pandas as pd
```

```python
# importing tensorflow and keras library for model building
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.activations import sigmoid , relu
```

```python
data=pd.read_csv('C:/Users/user/Downloads/Algo8Usecase/sample_data_intw.csv')
```

```python
# Removing unnecessary and contant features
data.drop(['msisdn','pdate','Unnamed: 0','pcircle'],axis=1,inplace=True)
```

```python
# Making all the reamining features numeric
cols=data.columns
for i in cols:
    data[i]=pd.to_numeric(data[i])
```

```python
# Filling NAN values
data['aon']=data['aon'].fillna(data['aon'].mode())
```

```python
# No. of data having label == 1
data['label'][data['label']==1].count()
```

```python
# No. of data having label == 0
data['label'][data['label']==0].count()
```

```python
# shuffling the dataset
data = data.sample(frac=1).reset_index(drop=True)
```

```python
# Appending column names in a list
col_names=[]
for i in data.columns:
    col_names.append(str(i))
col_names    
    
```

```python
target=data.label
```

```python
# scalling the dataset for better performance of model and reducing complexity
from sklearn.preprocessing import StandardScaler
scaller=StandardScaler()
```

```python
data2=data
```

```python
data2.drop('label',axis=1,inplace=True)
```

```python
data2=scaller.fit_transform(data2)
```

```python
data2=pd.DataFrame(data2,columns=col_names[1:])
```

```python
# Finding feature importance scores
from sklearn.ensemble import ExtraTreesClassifier as etc
x=data2.values
y=target.values
model=etc()
model.fit(x,y)
score=model.feature_importances_
score
```

```python
#appending feature names with their scores 
feature_importance=list(zip(data.columns,score))
from operator import itemgetter
feature_importance=sorted(feature_importance,key=itemgetter(1),reverse=True)
```

```python
feature_importance
```

```python
# applying a threshold value for selecting important features based on their scores
important_features=[x[0] for x in feature_importance if x[1]>0.005]
```

```python
# No. of important features selected 
len(important_features)
```

```python
df=pd.DataFrame()
```

```python
for i in important_features:
    df[i]=data2[i]
```

```python
df.head(5)
```

```python
len(target),len(df)
```

```python
df['label']=target
```

```python
cols=important_features
```

```python
# spliting dataset for train and test
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test= train_test_split(df[cols],df['label'], test_size=0.3)
```

```python
# As our dataset is highly imbalanced , so upsampling the minority data for training set to avoid the model bias
from sklearn.metrics import recall_score
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=44,ratio=1)
x_train, y_train = sm.fit_sample(x_train, y_train)
```

```python
len(x_train),len(y_train)
```

```python
# importing GridSearchCV and KerasClassifier for applying cross validation and to pick best no. of neurons,layers for my model
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
```

```python
# function for defining model graph

def create_model(layers,activation):
    model = Sequential()
    for i,nodes in enumerate(layers):
        if i==0:
            model.add(Dense(nodes,input_dim=x_train.shape[1]))
            model.add(Activation(activation))
        else:
            model.add(Dense(nodes))
            model.add(Activation(activation))
    model.add(Dense(1))
    model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
    return model
```

```python
model=KerasClassifier(build_fn=create_model,verbose=0)
```

```python
# parameters to be selected from for model training
layers=[[28,28],[40,30,]]

activations=['sigmoid','relu']

param_grid=dict(layers=layers,activation=activations,batch_size=[10],epochs=[100])

grid=GridSearchCV(estimator=model,param_grid=param_grid,cv=5)
```

```python
grid_result=grid.fit(x_train,y_train)
```

```python
[grid_result.best_score_,grid_result.best_params_]
```

```python
# Predicting test data (unseen data)
predicted=grid.predict(x_test)
```

```python
pred=[]
for i in predicted:
    if i>=0.5:
        pred.append(1)
    else:
        pred.append(0)
```

```python
# applying confusion matrix 
from sklearn.metrics import confusion_matrix
conf_mat = confusion_matrix(y_test, predicted)
print(conf_mat)
```

```python
# finding the  proportion of actual positives was identified correctly
recall=conf_mat[0][0]/(conf_mat[0][0]+conf_mat[1][0])
```

```python
# finding the proportion of positive identifications was actually correct
precision=conf_mat[0][0]/(conf_mat[0][0]+conf_mat[0][1])
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```
